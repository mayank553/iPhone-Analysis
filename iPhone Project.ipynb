```python
import pandas as pd
df = pd.read_csv('iphone.csv')
df
```

```python
df.columns
```

```python
#The column names have spaces . rename the column names to have underscore '_' instead of space (try to do in one go instead of specifying each column nam in rename method)

df.columns= df.columns.str.replace(' ', '_')
df.columns
```

```python
#star rating for some of the models is missing in the dataset. fill those missing values with the average rating all the models.

missing_ratings = df[df['Star_Rating'].isna()]
print(missing_ratings[['Product_Name', 'Star_Rating']])
```

```python
df['Star_Rating'].isna()
```

```python
df[df['Star_Rating'].isna()]
```

```python
df['Star_Rating'] = df['Star_Rating'].fillna(df['Star_Rating'].mean())
```

```python
df
```

```python
#df['Star_Rating'] = df.groupby('Ram')['Star_Rating'].transform(lambda x: x.fillna(x.mean()))
df['Star_Rating'].dtype

```

```python
# Now instead of filling missing values with avg rating of full dataset , fill with avg rating based on RAM. example :  if rating for a 2 gb phone is missing then take average of all other 2 gb phones rating and fill that value.
df['Star_Rating'] = df.groupby('Ram')['Star_Rating'].transform(lambda x: x.fillna(x.mean()))
df
```

```python
# create a new column in the dataframe "Discount_Percentage" based on MRP and sale value

df['Discount_Percentage'] = ((df['Mrp'] - df['Sale_Price']) / df['Mrp']) * 100
df
```

```python
# which model has highest percent discount ?

df = df.sort_values(by='Discount_Percentage', ascending=False)
df
```

```python
max_discount_model = df.loc[df['Discount_Percentage'].idxmax(), 'Product_Name']
max_discount_value = df['Discount_Percentage'].max()
max_discount_model
```

```python
# find total no of models each space configuration (128 GB , 64 GB etc)

df['Storage'] = df['Product_Name'].str.extract(r'(\d+\s*GB)')
storage_counts = df['Storage'].value_counts()
print(storage_counts)
```

```python
# find total number of models for each color 
import re

# Function to extract color from product name
def extract_color(product_name):
    match = re.search(r"\(([^,]+),", product_name)  # Captures everything before the comma
    return match.group(1).strip() if match else None  # Remove extra spaces if any

# Apply function to create a new column 'Color'
df['Color'] = df['Product_Name'].apply(extract_color)

# Display the updated DataFrame
print(df[['Product_Name', 'Color']].head())

```

```python
df
```

```python
#find total number of models for each color 
color_counts = df['Color'].value_counts()
print(color_counts)
```

```python
# list top 5 models having highest no of reviews 
df = df.sort_values(by='Number_Of_Reviews', ascending=False)
#df.head(5)
df[['Product_Name', 'Number_Of_Reviews']].head(5)
```

```python
# price diffrence between highest price and lowest price iphone (based on mrp)
price_difference = df['Mrp'].max() - df['Mrp'].min()
print("Price Difference:", price_difference)
```

```python
# find total no of reviews for iphone 11 and iphone 12 category . Output should have only 2 rows (for 11 and 12).

# Filter the DataFrame for iPhone 11 and iPhone 12 categories
iphone_11_reviews = df[df['Product_Name'].str.contains('iPhone 11', case=False)]
iphone_12_reviews = df[df['Product_Name'].str.contains('iPhone 12', case=False)]

# Calculate the total number of reviews for each category
total_reviews_11 = iphone_11_reviews['Number_Of_Reviews'].sum()
total_reviews_12 = iphone_12_reviews['Number_Of_Reviews'].sum()

# Create a new DataFrame with the results
result_df = pd.DataFrame({
    'iPhone Category': ['iPhone 11', 'iPhone 12'],
    'Total Reviews': [total_reviews_11, total_reviews_12]
})
```
